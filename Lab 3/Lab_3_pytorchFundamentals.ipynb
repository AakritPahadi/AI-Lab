## Overview of Common`torch` Tensor Creation Functions

   `torch.rand(shape)`: Creates a tensor with the specified `shape`and populates it with random numbers uniformly distributed between 0 and 1.
  `torch.zeros(shape)`: Creates a tensor with the specified `shape`and fills it with zeros.
  `torch.ones(shape)`: Creates a tensor with the specified `shape`and fills it with ones.
   `torch.eye(n)`: Creates a 2D identity matrix (a square matrix with ones on the main diagonal and zeros elsewhere) of size `n`x`n`.
`torch.arange(start, end, step)`: Creates a 1D tensor with values from `start` (inclusive) to `end`(exclusive), with a specified `step` size.
  `torch.linspace(start, end, steps)`: Creates a 1D tensor with `steps` number of equally spaced points between `start` and `end` (both inclusive).

```python
rand = torch.rand(3,4)
zeros = torch.zeros(3,4)
ones = torch.ones(3,4)
eye = torch.eye(3)
arange = torch.arange(start=0, end=10, step=2)
linspace = torch.linspace(start=0, end=10, steps=5)

print(f"rand: {rand}")
print(f"zeros: {zeros}")
print(f"ones: {ones}")
print(f"eye: {eye}")
print(f"arange: {arange}")
print(f"linspace: {linspace}")
```

# **Tensor Arithmetic Operations**

## Difference Between `+` Operator and `torch.add()`

Although `+` operator and `torch.add()` do element-wise additions on tensors, there is a difference between the two in respect to memory usage and broadcasting:

*  `+` operator: This is a Python overloaded operator that usually results in creation of a new tensor which holds the addition value. Using it may make your code more readable.

*  `torch.add(input, other, *, alpha=1, out=None)`: This is a function that gives you more options. Although like `+` it results in creation of a new tensor, it gives you the option to pass another tensor as `out` which will hold the resultant tensor. You do this to reuse memory and avoid creation of new memory blocks. There is also an `alpha` option to scale up the `other` tensor.

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([3, 4, 5])

print(a+b)
print(a-b)
print(a*b)

# adding using add()
print(torch.add(a,b))

print(torch.add(a, b, alpha=2))  # steps (a + bx)
```

## Difference Between `@` and `*` for Tensor Multiplication

For multiplication among `torch` tensors,

*   `*` (Element-Wise or Hadamard Multiplication): It does element-wise or Hadamard multiplication.It necessitates that both of them be having equal shapes or be broadcastable with respect to each other.The corresponding elements of both the tensors get multiplied together.

*   `@` (Matrix/Dot Multiplication): It involves performing matrix multiplication among tensors (or dot product in the case of 1D tensors).Certain conditions regarding the shapes of both the tensors arise.For instance, if two matrices `A`(m x n) and `B`(n x p) are there, then the resultant matrix `C` would have dimensions as (m x p).This is because of the condition that the no. of columns of the first matrix equals the no. of rows of the second matrix.

```python
# Element-wise multiplication (*)
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print(f"Tensor A: {a}")
print(f"Tensor B: {b}")
print(f"Element-wise multiplication (A * B): {a * b}")

# Matrix multiplication (@)
mat_a = torch.tensor([[1, 2],
                      [3, 4]])
mat_b = torch.tensor([[5, 6],
                      [7, 8]])

print(f"\nMatrix A: {mat_a}")
print(f"Matrix B: {mat_b}")
print(f"Matrix multiplication (A @ B): {mat_a @ mat_b}")

# Example with 1D tensors (dot product)
vec_a = torch.tensor([1, 2, 3])
vec_b = torch.tensor([4, 5, 6])

print(f"\nVector A: {vec_a}")
print(f"Vector B: {vec_b}")
print(f"Dot product (Vector A @ Vector B): {vec_a @ vec_b}")
```

### Understanding the Memory Behavior of `reshape()` and `view()`

The crucial distinction lies in whether the new tensor *shares memory* with the original tensor (a **view**) or if it creates a *completely new block of memory* (a **copy**).

*   **`tensor.view()`**: This method always attempts to return a **view**. This means the new tensor does not allocate new memory; it just presents the existing memory in a different shape. Because it shares memory, if you change an element in the `view` tensor, the original tensor will also change, and vice-versa. However, `view()` requires the original tensor to be **contiguous** in memory. If the original tensor is not contiguous (e.g., after transposing), `view()` will fail unless you call `.contiguous()` first, which *might* create a copy to make it contiguous.

*   **`tensor.reshape()`**: This method is more flexible. It will first try to return a **view** (sharing memory) if the memory layout allows it (i.e., if the tensor is contiguous and the new shape is compatible without reordering data). If the memory layout does not allow for a view (e.g., the tensor is not contiguous, or the new shape would require reordering elements that are not adjacent in memory), `reshape()` will automatically return a **copy** of the data in new memory. Changes to a copied tensor will *not* affect the original, and vice-versa.

**In summary:**
*   `view()`: Always a view (memory shared), but strict contiguity requirement.
*   `reshape()`: Prefers a view (memory shared) if possible, otherwise makes a copy (new memory). This makes `reshape()` generally safer and more convenient, as you don't have to worry about contiguity.

```python
t = torch.arange(1,10,1)  # numbers from 1-9 with step-size 1

print(t.reshape(3,3))  # converting t to 3x3 tensor
print(t.view(3,3))
print(t)
```

```python
x = torch.tensor([[1.], [2.], [3.]])
# squeeze(): Removes dimensions of size 1 from the tensor.
print(x.squeeze())
# unsqueeze(dim): Adds a dimension of size 1 at the specified position (dim).
print(x.unsqueeze(0))
```

```python
t1 = torch.tensor([1, 2, 3])
t2 = torch.tensor([4, 5, 6])
t3 = torch.tensor([7, 8, 9])
# torch.stack(): Concatenates a sequence of tensors along a new dimension.
# Here, dim=0 means a new dimension is added at the beginning, stacking t1 and t2 row-wise.
print(torch.stack([t1, t2, t3], dim=0))
print(torch.stack([t1, t2, t3], dim=1))
```

# **Indexing and Aggregation Operations**

```python
t = torch.tensor([[1.,2.,3.],[4.,5.,6.]])

print(t.min(), t.max())
print(t.mean(), t.sum())
print(t.argmin(), t.argmax())  # index of min/max

# indexing (like NumPy)
print(t[0])          # first row
print(t[:, 1])       # second column
print(t[1, 2])       # single element
```

# **Transferring Tensors to GPU**

## Why Are GPUs Preferred for Machine Learning?

**CPUs (Central Processing Units)** are designed for general-purpose computing. They have a few powerful cores optimized for complex sequential tasks and instruction-level parallelism. Think of a CPU as a **few highly skilled experts** who can tackle diverse problems one after another very quickly.

**GPUs (Graphics Processing Units)**, on the other hand, are specialized for parallel processing. They have thousands of smaller, more efficient cores that can perform many simple calculations simultaneously. Imagine a GPU as a **vast army of specialized workers** who can all perform the same simple task at the exact same time.

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

t = torch.tensor([1., 2., 3.])
t_gpu = t.to(device)
print(t_gpu.device)

# move back to CPU (needed before NumPy conversion)
t_cpu = t_gpu.cpu()
print(t_cpu.numpy())
```
